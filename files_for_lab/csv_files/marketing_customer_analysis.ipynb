```python
#Importing libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# The line down below is needed to prevent matplotlib to open the graph in a seaparate popup window. 
%matplotlib inline

#Importing CSV file

data = pd.read_csv('marketing_customer_analysis.csv')

data.head()
#data.shape
#data.columns
#data.dtypes

```

```python
# Countplot of the response column
sns.countplot(data=data, x='Response', palette= ['steelblue', 'lightskyblue'])

# Changing the plot title and axis labels
plt.title('Total Number of Responses')
plt.xlabel('Response')
plt.ylabel('Count')

# Show the graphic
plt.show()
```

```python
# Countplot of the response column
sns.countplot(data=data, x='Sales Channel', hue='Response', palette=['lightcoral', 'mediumaquamarine']) #exploring colors here

# Changing the plot title and axis labels
plt.title('Total Number of Responses per sales channels')
plt.xlabel('Channel')
plt.ylabel('Count')

# Show the graphic
plt.show()
```

```python
# Countplot of the response column
sns.barplot(data=data, x='Response', y='Total Claim Amount') 

# Changing the plot title and axis labels
plt.title('Response over Total Claim Amount')
plt.xlabel('Channel')
plt.ylabel('Count')

```

```python
# Countplot of the response column relative to income
sns.barplot(data=data, x='Response', y='Income', palette=['lightcoral', 'lightseagreen'])

# Changing the plot title and axis labels
plt.title('Responses relative to income')
plt.xlabel('Income')
plt.ylabel('Response')

# Show the graphic
plt.show()
```
